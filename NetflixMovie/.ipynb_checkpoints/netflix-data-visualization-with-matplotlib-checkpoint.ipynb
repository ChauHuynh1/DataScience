### load library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

### load data

```python
data=pd.read_csv('Data/NetflixMovie.csv')
data.head()
```

### missing values

There are several ways to process missing values. This time, we will fill it up, but drop it if there are two or more nan values in each row.

```python
data.isna().sum()
```

```python
data=data.dropna(thresh=10)
data=data.dropna(subset=['date_added','duration','rating'])
data=data.fillna('no data')
data=data.drop_duplicates()
```

### preprocessing

```python
data["date_added"]=pd.to_datetime(data['date_added'])

data['month_added']=data['date_added'].dt.month
data['month_name_added']=data['date_added'].dt.month_name()
data['year_added']=data['date_added'].dt.year

data.head()
```

### palette

```python
sns.palplot(['#221f1f', '#b20710', '#e50914','#f5f5f1'])

plt.title("my palette ",loc='left',fontfamily='serif',fontsize=15,y=1.2)
plt.show()
```

```python
# https://www.businessinsider.com/how-netflix-has-looked-over-the-years-2016-4

tl_dates=[
    "1997\nBegin",
    "1998\nDVD&Mail Service",
    "2003\nPublic",
    "2007\nStreaming service",
    "2016\nGlobal",
    "2021\nRight now, Netflix."
]

tl_x=[1, 2, 4, 5.5, 8,9]

fig, ax = plt.subplots(figsize=(15, 4))
background_color = '#F2F2F2'
fig.patch.set_facecolor(background_color) 
ax.set_facecolor(background_color) 

ax.set_ylim(-2, 1.75)
ax.set_xlim(0, 10)

ax.axhline(0, xmin=0.1, xmax=0.9, c='#e50914', zorder=1)

ax.scatter(tl_x, np.zeros(len(tl_x)), s=120, c='#f5f5f1', zorder=2)
ax.scatter(tl_x, np.zeros(len(tl_x)), s=30, c='#221f1f', zorder=3)

for x, date in zip(tl_x, tl_dates):
    ax.text(x, -0.55, date, ha='center', 
            fontfamily='serif', fontweight='bold',
            color='#4a4a4a',fontsize=11)
    
for spine in ["left", "top", "right", "bottom"]:
    ax.spines[spine].set_visible(False)

    ax.set_xticks([]) 
ax.set_yticks([]) 

ax.set_title("Netflix through the years", fontweight="bold", fontfamily='serif', fontsize=16, color='#221f1f')
ax.text(1.2,1.57,"Netflix's website in 1999 looked nothing like it does today  here's how it has evolved over the years", fontfamily='serif', fontsize=12, color='#221f1f')

plt.show()
```

```python
df=pd.DataFrame(data['type'].value_counts(normalize=True)).T

fig, ax = plt.subplots(1,1,figsize=(6.5, 2.5),dpi=200)

fig.patch.set_facecolor('#F2F2F2')
ax.set_facecolor('#F2F2F2')
ax.barh(df.index, df['Movie'], 
        color='#b20710', alpha=0.9, label='Movie')
ax.barh(df.index, df['TV Show'], left=df['Movie'], 
        color='#221f1f', alpha=0.9, label='TV Show')

for i in df.index:
    ax.annotate(f"{int(df['Movie'][i]*100)}%", 
                   xy=(df['Movie'][i]/2, i),
                   va = 'center', ha='center',fontsize=40, fontweight='light', fontfamily='serif',
                   color='white')

    ax.annotate("Movie", 
                   xy=(df['Movie'][i]/2, -0.25),
                   va = 'center', ha='center',fontsize=15, fontweight='light', fontfamily='serif',
                   color='white')
    
for i in df.index:
    ax.annotate(f"{int(df['TV Show'][i]*100)}%", 
                   xy=(df['Movie'][i]+df['TV Show'][i]/2, i),
                   va = 'center', ha='center',fontsize=40, fontweight='light', fontfamily='serif',
                   color='white')
    ax.annotate("TV Show", 
                   xy=(df['Movie'][i]+df['TV Show'][i]/2, -0.25),
                   va = 'center', ha='center',fontsize=15, fontweight='light', fontfamily='serif',
                   color='white')

fig.text(0.125,1.03,'Movie & TV Show distribution', fontfamily='serif',fontsize=15, fontweight='bold')
fig.text(0.125,0.92,'There are more Movie than TV shows on Netflix.',fontfamily='serif',fontweight='light',fontsize=12)  

for s in ['top', 'left', 'right', 'bottom']:
    ax.spines[s].set_visible(False)
    
ax.legend().set_visible(False)
ax.set_xticks([])
ax.set_yticks([])
plt.show()
```

```python
df=data['country'].value_counts()[:11].drop('no data',axis=0)
df.index=['USA','India','UK','Japan','S.Korea','Canada','Spain','France','Mexico','Egypt']

color_map=['#221f1f' for _ in range(10)]
color_map[0]=color_map[1]='#e50914'

fig=plt.figure(dpi=200)
gs=fig.add_gridspec(1,1)
ax=fig.add_subplot(gs[0,0])
fig.patch.set_facecolor('#F2F2F2')
ax.set_facecolor('#F2F2F2')
ax.bar(df.index,df,zorder=2,color=color_map)

for i in df.index:
    m=df.loc[i]
    ax.annotate(f'{m}',
               xy=(i,m+150),va='center',ha='center',fontweight='light',fontfamily='serif',
               fontsize=8)

ax.set_ylim(0,3500)
ax.grid(axis='y',linestyle=':',alpha=0.4,zorder=1)
ax.set_xticks(range(10))
ax.set_xticklabels(df.index,fontsize=6,fontfamily='serif',fontweight='light')
for s in ['top','right','left']:
    ax.spines[s].set_visible(False)

ax.tick_params(which='both',axis='both',length=0)

fig.text(0.152, 1.01, 'Top 10 countries on Netflix', fontsize=15, fontweight='bold', fontfamily='serif')
fig.text(0.152, 0.95, 'The three most frequent countries have been highlighted.', fontsize=12, fontweight='light', fontfamily='serif')

fig.text(1.1, 1.01, 'Insight', fontsize=15, fontweight='bold', fontfamily='serif')

fig.text(1.1, 0.84, '''
The United States and India have the largest number
and Korea ranks fifth.
''', fontsize=12, fontweight='light', fontfamily='serif')

import matplotlib.lines as lines
l1 = lines.Line2D([1, 1], [0, 1], transform=fig.transFigure, figure=fig,color='black',lw=0.2)
fig.lines.extend([l1])

plt.show()
```

```python
idx=data['country'].value_counts()[:11].drop('no data',axis=0).index
df=data.groupby('country')['type'].value_counts(normalize=True).unstack().loc[idx]
df=df.sort_values(by='Movie',ascending=True)

fig, ax = plt.subplots(1,1,figsize=(15, 8),dpi=200)
ax.set_facecolor('#F2F2F2')
fig.patch.set_facecolor('#F2F2F2')
ax.barh(df.index, df['Movie'], 
        color='#b20710', alpha=0.8, label='Movie')
ax.barh(df.index, df['TV Show'], left=df['Movie'], 
        color='#221f1f', alpha=0.8, label='TV Show')

for i in df.index:
    m=df['Movie']
    n=df['TV Show']
    ax.annotate(f'{m[i]:.2%}',xy=(m[i]/2,i),fontweight='light',fontfamily='serif',fontsize=9,color='white',
               va='center',ha='center')
    ax.annotate(f'{n[i]:.2%}',xy=(m[i]+n[i]/2,i),fontweight='light',fontfamily='serif',fontsize=9,color='white',
               va='center',ha='center')
    
for s in ['top','right','bottom','left']:
    ax.spines[s].set_visible(False)
ax.tick_params(which='both',axis='both',length=0)
ax.set_xticks([])
ax.set_yticklabels(df.index,fontweight='bold',fontfamily='serif',fontsize=10)

fig.text(0.13, 0.93, 'Top 10 countries Movie & TV Show split', fontsize=15, fontweight='bold', fontfamily='serif')   
fig.text(0.131, 0.89, 'Percent Stacked Bar Chart', fontsize=12,fontfamily='serif')   

fig.text(0.75,0.9,"Movie", fontweight="bold", fontfamily='serif', fontsize=15, color='#b20710')
fig.text(0.805,0.9,"|", fontweight="bold", fontfamily='serif', fontsize=15, color='#221f1f')
fig.text(0.82,0.9,"TV Show", fontweight="bold", fontfamily='serif', fontsize=15, color='#221f1f')

fig.text(0.95, 0.92, 'Insight', fontsize=15, fontweight='bold', fontfamily='serif')

fig.text(0.95, 0.86, '''
India has a large proportion of movies, and Korea has a low proportion of movies.
''', fontsize=12, fontweight='light', fontfamily='serif')

l1 = lines.Line2D([0.93,0.93],[0.1,0.9], transform=fig.transFigure, figure=fig,color='black',lw=0.2)
fig.lines.extend([l1])
plt.show()
```

```python
us=data[data['country']=='United States']['year_added'].value_counts().sort_index()
kor=data[data['country']=='South Korea']['year_added'].value_counts().sort_index()

fig, ax = plt.subplots(1, 1, figsize=(12, 6),dpi=200)
color = ["#b20710", "#221f1f"]

for i, country in enumerate([us,kor]):
    ax.fill_between(country.index, 0, country, color=color[i], label=country,alpha=0.9,zorder=i)
      
ax.yaxis.tick_right()

for s in ['top', 'right','bottom','left']:
    ax.spines[s].set_visible(False)

ax.grid(False)

ax.set_xlim(2008,2020)
plt.xticks(np.arange(2008, 2021, 1))

fig.text(0.13, 0.85, 'Movies & TV Shows added over time', fontsize=15, fontweight='bold', fontfamily='serif')
fig.text(0.13, 0.58, 
'''Netflix peak global content amount was in 2019.

Few contents have been added as they have slowed down due to COVID-19.
When looking at the difference in content
between the United States and Korea, it is severe.'''
, fontsize=12, fontweight='light', fontfamily='serif')


fig.text(0.13,0.2,"USA", fontweight="bold", fontfamily='serif', fontsize=15, color='#b20710')
fig.text(0.19,0.2,"|", fontweight="bold", fontfamily='serif', fontsize=15, color='black')
fig.text(0.2,0.2,"S.Korea", fontweight="bold", fontfamily='serif', fontsize=15, color='#221f1f')

ax.tick_params(axis='both', which='both',length=0)

plt.show()
```
